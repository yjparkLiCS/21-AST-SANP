```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
DEM_DATA_FILENMAE = './DEM_data/apollo.npy'
RECON_FILENAME = './results/recon_6e5ffe3110e44ea1a6dde5fed355ff29.npz'
```

```python
dem_data = np.load(DEM_DATA_FILENMAE)
recon_data = np.load(RECON_FILENAME)
```

### Original DEM

```python
plt.figure()
plt.imshow(dem_data)
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.axis('equal')
plt.tight_layout() 
plt.title('Original DEM')
plt.show()
```

### Reconstructed DEM

```python
dem_data_recon = dem_data.copy()
dem_data_recon[recon_data['idx'][:, 0], recon_data['idx'][:, 1]] = recon_data['y_pred']
```

```python
plt.figure()
plt.imshow(dem_data_recon)
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.axis('equal')
plt.tight_layout() 
plt.title('Reconstructed DEM')
plt.show()
```
